# Defining a Negative Triangularity plasma using IETy

In this notebook we define a negative triangularity plasma using **IETy** starting from a real plasma shot of RFX-mod experiment. 

The steps to follow are the following:
* **Plasma boundary shape definition** 
    * Loading of the original plasma boundary shape of RFX-mod experimental shot no. 36922;
    * Changing the shape in order to get negative triangularity
    * Defining the mesh for the solution of equilibrium problem
* **Solving fixed boundary plasma equilibrium problem**
    * Defining the total plasma current, plasma current density profile and poloidal flux at boundary
    * Solving the equilibrium problem
    
* **Computing the coil currents needed for such equilibrium**
    * Evaluating the coils contribution to poloidal flux at boundary
    * Solving optimization/minimization problem

# Plasma boundary shape definition

Firstly the original plasma shape is loaded; then the four extremal points of the boundary are found and used to compute all the plasma shape parameters.

```python
import numpy as np
from matplotlib import pyplot as plt
from funShaping import *
import dmsh
import optimesh
import meshplex
import networkx as nx # For ordering boundary nodes
import scipy.io

input_data = scipy.io.loadmat('36922BoundaryGUI')
geomBoundary = input_data["geometry"]["boundaryCoordinates"][0][0]
R_boundary_ref = geomBoundary[:,0]
Z_boundary_ref = geomBoundary[:,1]

router = np.max(R_boundary_ref)
zouter = Z_boundary_ref[np.argmax(R_boundary_ref)]
rinner = np.min(R_boundary_ref)
zinner = Z_boundary_ref[np.argmin(R_boundary_ref)]
rupper = R_boundary_ref[np.argmax(Z_boundary_ref)]
zupper = np.max(Z_boundary_ref)
rlower = R_boundary_ref[np.argmin(Z_boundary_ref)]
zlower = np.min(Z_boundary_ref)
Boundary = {"router":router, "zouter":zouter, 
            "rinner":rinner, "zinner":zinner, 
            "rupper":rupper, "zupper":zupper, 
            "rlower":rlower, "zlower":zlower}

shape_parameters = compute_shape_parameters(Boundary)
print('===========ORIGINAL BOUNDARY===========')
print("(R,Z) outer", (Boundary["router"],Boundary["zouter"]))
print("(R,Z) inner", (Boundary["rinner"],Boundary["zinner"]))
print("(R,Z) upper", (Boundary["rupper"],Boundary["zupper"]))
print("(R,Z) lower", (Boundary["rlower"],Boundary["zlower"]))

print("Major radius (m): ", shape_parameters["major_radius"])
print("Minor radius (m): ", shape_parameters["minor_radius"])
print("elongation: ", shape_parameters["elongation"])
print("upper elongation: ", shape_parameters["upper_elongation"])
print("lower elongation: ", shape_parameters["lower_elongation"])
print("orizontal elongation: ", shape_parameters["horizontal_elongation"])
print("triangularity: ", shape_parameters["triangularity"])
print("upper triangularity: ", shape_parameters["upper_triangularity"])
print("lower triangularity: ", shape_parameters["lower_triangularity"])
```

Now we change the triangularity of such plasma from positive to negative. Thus, it is needed to change the radial coordinate of the upper and lower point (*rupper, rlower*) in order to make it larger than the geometrical center of the plasma boundary. Then, the new boundary is obtained by means of four branches of superellipses. The shape of the plasma can be modified without affecting the main shape parameters by changing the squareness. 

```python
Boundary = {"router":router, "zouter":zouter, 
            "rinner":rinner, "zinner":zinner, 
            "rupper":2.2, "zupper":zupper, 
            "rlower":2.2, "zlower":zlower}
shape_parameters = compute_shape_parameters(Boundary)

print('===========NEGATIVE TRIANGULARITY BOUNDARY===========')
print("(R,Z) outer", (Boundary["router"],Boundary["zouter"]))
print("(R,Z) inner", (Boundary["rinner"],Boundary["zinner"]))
print("(R,Z) upper", (Boundary["rupper"],Boundary["zupper"]))
print("(R,Z) lower", (Boundary["rlower"],Boundary["zlower"]))

print("Major radius (m): ", shape_parameters["major_radius"])
print("Minor radius (m): ", shape_parameters["minor_radius"])
print("elongation: ", shape_parameters["elongation"])
print("upper elongation: ", shape_parameters["upper_elongation"])
print("lower elongation: ", shape_parameters["lower_elongation"])
print("orizontal elongation: ", shape_parameters["horizontal_elongation"])
print("triangularity: ", shape_parameters["triangularity"])
print("upper triangularity: ", shape_parameters["upper_triangularity"])
print("lower triangularity: ", shape_parameters["lower_triangularity"])

r_ellipse,z_ellipse,k,A,B = computeEllipseBoundary(Boundary,shape_parameters)

print('===========NEW SQUARENESS===========')
csi = np.zeros([4,1]) 
csi[0,0]=0
csi[1,0]=-0.3
csi[2,0]=0.2
csi[3,0]=0

x,y = computeSuperEllipses(Boundary,shape_parameters,csi)
xf = np.array([x[0,:-1],x[1,:-1],x[2,:-1],x[3,:-1]]).flatten()
yf = np.array([y[0,:-1],y[1,:-1],y[2,:-1],y[3,:-1]]).flatten()
xplot = np.r_[xf, xf[0]]
yplot = np.r_[yf, yf[0]]

# plot the result
fig,ax = plt.subplots(1,3,figsize = (8,4), sharex = False)
ax[0].plot(R_boundary_ref,Z_boundary_ref,'k')
ax[0].axis('equal')
ax[0].grid()
ax[0].set_title('ORIGINAL BOUNDARY')

ax[1].plot(R_boundary_ref,Z_boundary_ref,'k',linewidth=0.5)
ax[1].plot(router, zouter, 'o')
ax[1].plot(rinner, zinner, 'o')
ax[1].plot(rupper, zupper, 'o')
ax[1].plot(rlower, zlower, 'o')
ax[1].plot(r_ellipse, z_ellipse, 'b.')
ax[1].axis('equal')
ax[1].grid('minor')
ax[1].set_title('NEGATIVE TRIANGULARITY')

ax[2].plot(router, zouter, 'o')
ax[2].plot(rinner, zinner, 'o')
ax[2].plot(rupper, zupper, 'o')
ax[2].plot(rlower, zlower, 'o')
ax[2].plot(R_boundary_ref, Z_boundary_ref, 'k',linewidth=0.5)
ax[2].plot(r_ellipse, z_ellipse, 'b.',alpha=.2)
ax[2].plot(xplot,yplot,'r-',linewidth=2)
ax[2].axis('equal')
ax[2].grid('minor')
ax[2].set_title('NEW SQUARENESS')
```

Now that the plasma boundary is defined, the mesh of the plasma domain can be generated. 

```python
print('================Defining primal mesh================')
R = xf[0::2].T
Z = yf[0::2].T
coordinates = np.array([R.flatten(), Z.flatten()]).T
coordinates_list = coordinates.tolist()
geo = dmsh.Polygon(coordinates_list) 

#geo.show()
p, t = dmsh.generate(geo, target_edge_length=0.05)
p, t = optimesh.optimize_points_cells(p, t, "CVT (full)", 1.0e-5, 100)
#dmsh.helpers.show(p, t, geo)

print('================Finding boundary elements of primal mesh================')
# ============================================
# Generate mesh object
mesh = meshplex.MeshTri(p, t)

# ============================================  
# Get boundary edge indices
bei = mesh.boundary_edges    # boundary edge indicies
et  = mesh.edges['points']   # total edges
eb  = et[bei,:]              # boundary edges

# ============================================
# Get boundary nodes - unordered
b = list({e for a in eb for e in a})

print('================Computing Dual Mesh================')
Area_duale, kk_nu = compute_dual_mesh_line(p,t)

plt.figure(figsize = (4,2))
plt.triplot(p[:,0], p[:,1], t)
plt.grid()
plt.axis('equal')
```

# Solving fixed boundary plasma equilibrium problem

The equilibrium is defined by specifying the total plasma current, flux at boundary and plasma current density profile. Then it is solved with fixed-point-Picard scheme.

```python
from funEquil import *
#from scipy.interpolate import griddata

Ipla = 50e3
psiBoundary = 0.7
R0=     1.995
Beta0= 0.3
AlphaM= 1.5
AlphaN= 1.001
tol = 1e-12
ngrid = 100
mu0 = 4*np.pi*1e-7
nn = p.shape[0]
ntri = t.shape[0]
nfaces = t.shape[1]
nodi_BCs= b
Iphi = np.zeros((nn,1))

Areatot = sum(Area_duale)

Jphi = Ipla/Areatot
Iphi = Jphi * Area_duale
Lambda = Ipla/sum(Iphi)
Iphi=Iphi*Lambda

print('Imposing boundary conditions BCs')
Psi = np.zeros((nn,1))
solk = np.zeros((nn,1))

kk_bc = np.zeros((nn,nn))
kk_bc = np.copy(kk_nu)
nBCs = len(nodi_BCs)
Psi[nodi_BCs] = psiBoundary

kk_bc[nodi_BCs,:]=0
kk_bc[:,nodi_BCs]=0
for ii in range(nBCs):
    kk_bc[nodi_BCs[ii],nodi_BCs[ii]]=1

Iphi = Iphi - np.dot(kk_nu[:,nodi_BCs] ,Psi[nodi_BCs])
Iphi[nodi_BCs] = Psi[nodi_BCs]

#first solution
sol_0 = np.linalg.solve(kk_bc, Iphi)
# sol_0 = np.linalg.lstsq(kk_bc, Iphi)
# plt.plot(sol_0,'.')

solk=np.copy(sol_0)#.flatten()
psiAxis0 = np.max(solk) # magnetic axis

#post processing: magnetic axis, contours of initial solution
rr =  np.zeros((nn,1))
zz =  np.zeros((nn,1))
rr[:,0] = p[:,0]
zz[:,0] = p[:,1]

rgrid = np.linspace(min(rr),max(rr),ngrid)
zgrid = np.linspace(min(zz),max(zz),ngrid)
RR,ZZ = np.meshgrid(rgrid,zgrid)

psi_ak = max(solk) #%finding magnetic axis
errorAxis = np.zeros((nn,1))
errorAxis[0]=(psi_ak - psiAxis0)/abs(psiAxis0)*100 #%per cent relative magnetic axis error

fig,ax = plt.subplots(figsize = (4,2), sharex = False)
ax.set_aspect('equal', adjustable='box')
surf=ax.tricontourf(p[:,0], p[:,1],t, solk.flatten(),50,alpha=.35)
ax.tricontour(p[:,0], p[:,1],t, solk.flatten(),10,linewidths=2)
fig.colorbar(surf,ax = ax,orientation='vertical')
ax.set_title('First solution')

#iterative procedure solution for the magnetic poloidal flux
n_iter = 0
err = 1
scarto = np.zeros((nn,1))
Residual_Picard = np.zeros((nn,1))
Iphi_plasma = np.zeros((nn,1))
Lambda=1

while err > tol:
    psin = (solk-psi_ak)/(psiBoundary-psi_ak)# %normalized poloidal magnetic flux
    g_psi=(1-psin**AlphaM)**AlphaN
    h_f = rr*Beta0/R0+R0*(1-Beta0)/rr
    
    Jphi = np.zeros((nn,1))
    Jphi = Lambda*np.multiply(h_f,g_psi)

    Iphi = np.zeros((nn,1))
    Iphi = np.multiply(Area_duale,Jphi)
    Lambda = Ipla/sum(Iphi) #calculation of lambda and total current
    Iphi = Iphi*Lambda
    Iphi_tot = sum(Iphi)
    Iphi_plasma = np.copy(Iphi)
    
    Psi[nodi_BCs] = psiBoundary
    Iphi=Iphi-np.dot(kk_nu[:,nodi_BCs],Psi[nodi_BCs])
    Iphi[nodi_BCs]=Psi[nodi_BCs]
    solk1 = np.linalg.solve(kk_bc, Iphi)
    
    #solution
    psi_aknew = solk1.max()#%finding magnetic axis
    ind_psiak = np.argmax(solk1)
    errorAxis[n_iter] = abs((psi_aknew-psi_ak)/psi_ak)*100 #%calculation of error at axis
    psi_ak = psi_aknew
    
    #calculation of iterative error (Picard)
    err = np.linalg.norm(solk1-solk)
    scarto[n_iter,:] = np.linalg.norm(solk1-solk)/len(solk1)
         
    if n_iter==0:
        print('---------------------------')          
        print('#iter: ', n_iter+1)
        print('lambda: ',Lambda)
        print('I tot: ',Iphi_tot)
        print('Psi magnetic axis: ',psi_ak)
        print('R(m) magnetic axis: ',rr[ind_psiak])
        print('Z(m) magnetic axis: ',zz[ind_psiak])
        print('Psi error at axis: ',errorAxis[n_iter])
        print('Convergence error: ',err)
        
    
    # re-assignement of iterative solution        
    solk = np.copy(solk1)
    Residual_Picard[n_iter,:]=err
    n_iter = n_iter+1    
            
    if n_iter>50:        
        print('------EXCEEDED 50 ITERATIONS--------------------')
        break

print('---------------------------')          
print('#iter: ', n_iter+1)
print('lambda: ',Lambda)
print('I tot: ',Iphi_tot)
print('Psi magnetic axis: ',psi_ak)
print('R(m) magnetic axis: ',rr[ind_psiak])
print('Z(m) magnetic axis: ',zz[ind_psiak])
print('Psi error at axis: ',errorAxis[n_iter])
print('Convergence error: ',err)
Psi_final = np.zeros((nn,1))

Jphi_final = Iphi_plasma/Area_duale
Psi_final = np.copy(solk)
Raxis = rr[ind_psiak]
Zaxis = zz[ind_psiak]
Psi_axis = psi_ak
psiBoundary = Psi_final.min()

fig,ax = plt.subplots(1,2,figsize = (6,3), sharex = False)
ax[0].set_aspect('equal', adjustable='box')
ax[1].set_aspect('equal', adjustable='box')
cont=ax[0].tricontourf(p[:,0], p[:,1],t, Psi_final.flatten(),50,alpha=.35)
ax[0].tricontour(p[:,0], p[:,1],t, Psi_final.flatten(),10,linewidths=2)
fig.colorbar(cont,ax = ax[0],orientation='vertical')
ax[0].set_title("$\Psi$")

surfJ=ax[1].tricontourf(p[:,0], p[:,1],t, Jphi_final.flatten(),50,alpha=1,cmap=plt.get_cmap('jet'))
# ax[1].tricontour(p[:,0], p[:,1],t, Jphi_final.flatten(),50,linewidths=2,cmap=plt.get_cmap('jet'))
fig.colorbar(surfJ,ax = ax[1],orientation='vertical')
ax[1].set_title("$J_\phi$")
fig.suptitle('Final solution')
fig.tight_layout()
```

# Computing the coil currents needed for such equilibrium

Here the coil currents needed to obtain such shape are computed.

```python
import warnings
warnings.filterwarnings("ignore")
def plotPsimap(X,Y,Z,ax):
    # Clear the current plot figure
    ax.clear()
    surf = ax.contourf(X, Y, Z,50,   cmap=plt.get_cmap('viridis'),alpha=.45)
    cont = ax.contour(X, Y, Z,10,cmap=plt.get_cmap('viridis'),linewidths=2)  
    xw = 1.995+0.46*np.cos(np.linspace(0,2*np.pi,100))
    yw = 0+0.46*np.sin(np.linspace(0,2*np.pi,100))
    ax.plot(xw,yw,color="black",linewidth=1)
    # ax.set_axis_off()
    ax.set_xlim([X.min(), X.max()])
    ax.set_ylim([Y.min(), Y.max()])
    # ax.set_title('{:.4f} ms'.format(k*dt*1000))
    # ax.axis('equal') 
    return surf,cont

def plotJmap(X,Y,Z,ax):
    # Clear the current plot figure
    ax.clear()
    surf = ax.contourf(X, Y, Z,50,   cmap=plt.get_cmap('jet'),alpha=.45)
    cont = ax.contour(X, Y, Z,10,cmap=plt.get_cmap('jet'),linewidths=1)  
    xw = 1.995+0.46*np.cos(np.linspace(0,2*np.pi,100))
    yw = 0+0.46*np.sin(np.linspace(0,2*np.pi,100))
    ax.plot(xw,yw,color="black",linewidth=1)
    # ax.set_axis_off()
    ax.set_xlim([X.min(), X.max()])
    ax.set_ylim([Y.min(), Y.max()])
    # ax.set_title('{:.4f} ms'.format(k*dt*1000))
    # ax.axis('equal') 
    return surf,cont

from scipy.interpolate import griddata

Psi_grid_final = griddata(p,Psi_final,(RR,ZZ), method='cubic')
Psi_grid_final= Psi_grid_final[:,:,0]
Jphi_grid_final = griddata(p,Jphi_final,(RR,ZZ), method='cubic')
Jphi_grid_final= Jphi_grid_final[:,:,0]

print('-----------------------------------------------------------------------')
print("-------------------------PREPROC - OPTIMIZATION------------------------")
print('-----------------------------------------------------------------------')
#########################################
# Triangle point indices
# Find centroids of triangles
rbar=(rr[t[:,0]]+rr[t[:,1]]+rr[t[:,2]])/3
zbar=(zz[t[:,0]]+zz[t[:,1]]+zz[t[:,2]])/3

#Calculating plasma boundary flux contribution 
#interpolating J over the triangles centroids
Jp_b = griddata(p,Jphi_final,(rbar,zbar), method='cubic')
Jp_b = Jp_b[:,:,0]
def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))

Area_primale = np.zeros((ntri,1))
for i in range(ntri):
    x = rr[t[i,:]].flatten()
    y = zz[t[i,:]].flatten()
    Area_primale[i]=PolyArea(x,y)

Ipla = np.multiply(Jp_b,Area_primale) # % calculating total current per element

#selecting only boundary points
print("-------------------plasma boundary points")
Rpoints = rr[nodi_BCs]
Zpoints = zz[nodi_BCs]
points = {"R":Rpoints, "Z":Zpoints}
nr = len(Rpoints)
nz = len(Zpoints)

print("-------------------Plasma Green Matrix")
Rpla = rbar
Zpla = zbar
plasma = {"R":Rpla,"Z":Zpla}
G_a, G_br, G_bz, G_psi = fun_Field_Loop(plasma, points)

Psi_pla = np.dot(G_psi,Ipla)
Psi_coils = Psi_final[nodi_BCs] - Psi_pla


print("-------------------Defining coils")
input_data = scipy.io.loadmat('inputTestIETy')
Rcoil = input_data["coils"]['R'][0][0][0]
Zcoil = input_data["coils"]['Z'][0][0][0]
DR = input_data["coils"]['DR'][0][0][0]
DZ = input_data["coils"]['DZ'][0][0][0]
Nturns = input_data["coils"]['turns'][0][0][0]
current = np.ones(56)

kon = input_data["coils"]['kon'][0][0]
kon = kon

source = {"R":Rcoil, "Z":Zcoil, "current":current}

## optimization
# print("-------------------plasma boundary points")
# Rpoints = input_data["space"]['RR'][0][0]
# Zpoints = input_data["space"]['ZZ'][0][0]
# points = {"R":Rpoints, "Z":Zpoints}
# nr = len(Rpoints)
# nz = len(Zpoints)

# print("-------------------Plasma Green Matrix")
# Rpla = input_data["plasma"]['R'][0][0]
# Zpla = input_data["plasma"]['Z'][0][0]
# Ipla = input_data["plasma"]['current'][0][0]
# plasma = {"R":Rpla,"Z":Zpla}
# G_a, G_br, G_bz, G_psi = fun_Field_Loop(plasma, points)

# Psi_pla = np.dot(G_psi,Ipla)
# Psi_tot =  input_data["psi_tot"]
# Psi_coils = Psi_tot - Psi_pla

print("-------------------PREPROC - OPTIMIZATION-------------------")
print("-------------------Coils Green Matrix")
source = {"R":Rcoil, "Z":Zcoil,"DR":DR,"DZ":DZ,"current":current}
NGauss = 2
[G_ac,G_brc,G_bzc,G_psic]=fun_Field_Coil( source, points, NGauss)
G_ac = np.multiply(G_ac,Nturns)
G_brc = np.multiply(G_brc,Nturns)
G_bzc = np.multiply(G_bzc,Nturns)
G_psic = np.multiply(G_psic,Nturns)

G_psic = np.dot(G_psic,kon.T)

print("-------------------OPTIMIZATION-------------------")
print("-------------------LEAST SQUARE MINIMIZATION")
# LEAST SQUARE MINIMIZATION
# minimize 0.5 * ||A x - b||**2
# subject to lb <= x <= ub
A = G_psic
b = Psi_coils.flatten()

ImaxMW = 50e3
ImaxFSW = 6.25e3
ub = np.concatenate((ImaxMW*np.ones(4),  ImaxFSW*np.ones(4), [np.finfo(float).eps], [ImaxFSW], [np.finfo(float).eps], ImaxFSW*np.ones(3)))
lb = np.concatenate((-ImaxMW*np.ones(4),  -ImaxFSW*np.ones(4), [-np.finfo(float).eps], [-ImaxFSW], [-np.finfo(float).eps], -ImaxFSW*np.ones(3)))
from scipy.optimize import lsq_linear
res_ls = lsq_linear(A, b, bounds=(lb, ub), lsmr_tol='auto', verbose=1)

print("-------------------NL MINIMIZATION")
def OF(x):
    # np.sqrt(sum((np.dot(A,x)-b)**2)/len(b))
    objF = 0.5*np.sqrt(np.mean(np.square(np.dot(A,x)-b)))
    return objF

# global SLSQP
from scipy.optimize import Bounds, minimize
bounds = Bounds(lb, ub)
x0 = 1e3*np.ones(14)
res_min = minimize(OF, x0, method='trust-constr', 
                  tol= 1e-12,
                  bounds=bounds)
print("-------------------PLOT-------------------")
nr, nz = (100, 100)
r = np.linspace(1.4, 2.5, nr)
z = np.linspace(-0.55, 0.55, nz)
rv, zv = np.meshgrid(r, z)
space_plot = {"R":rv, "Z":zv} 

# plasma
G_pla_psi_plot = fun_Field_Loop(plasma, space_plot)[3]

Psi_pla_plot = np.dot(G_pla_psi_plot,Ipla)
Psi_pla_plot = np.reshape(Psi_pla_plot, (nr, nz),order="F")

# # coils
G_c_psi_plot=fun_Field_Coil( source, space_plot, NGauss)[3]
GG_Psi_c_plot= np.multiply(G_c_psi_plot,Nturns)

I_ls = np.dot(res_ls.x,kon)
Psi_c_plot_ls = np.dot(GG_Psi_c_plot,I_ls)
Psi_c_plot_ls = np.reshape(Psi_c_plot_ls, (nr, nz),order="F")

I_min = np.dot(res_min.x,kon)
Psi_c_plot_min = np.dot(GG_Psi_c_plot,I_min)
Psi_c_plot_min = np.reshape(Psi_c_plot_min, (nr, nz),order="F")

Psi_ls = Psi_c_plot_ls+Psi_pla_plot
Psi_trustcon = Psi_c_plot_min+Psi_pla_plot

##########
fig,ax = plt.subplots(1,2,figsize = (6,3), sharex = False)
ax[0].set_aspect('equal', adjustable='box')
ax[1].set_aspect('equal', adjustable='box')
[surf,cont] = plotPsimap(rv,zv,Psi_ls,ax[0])
# fig.colorbar(cont,ax = ax[0],orientation='horizontal')
fig.colorbar(surf,ax = ax[0],orientation='horizontal')
ax[0].set_title("least-square")
[surf,cont] = plotPsimap(rv,zv,Psi_trustcon,ax[1])
fig.colorbar(surf,ax = ax[1],orientation='horizontal')
ax[1].set_title("trust-constr")
fig.tight_layout()

#plt.figure(figsize = (2,2))
#plt.plot(Psi_ls.flatten(),'k',label="least-square")
#plt.plot(Psi_trustcon.flatten(),'r',label="trust-constr")
#plt.grid()
#plt.legend()

print("RMSE lsqlin  :",OF(res_ls.x))
print("RMSE minimize:",OF(res_min.x))

width=0.4
plt.figure(figsize = (6,3))
plt.bar(range(len(res_ls.x)),res_ls.x/1e3,width=width)
plt.bar(np.arange(len(res_min.x))+width,res_min.x/1e3,width=width)
plt.grid(color='#95a5a6', linestyle='--', linewidth=1, axis='both', alpha=0.7)
plt.ylabel('Coil currents (kA)')
plt.xticks(range(len(res_ls.x)), ['MW1', 'MW2', 'MW3','MW4',  'FSW1','FSW2','FSW3',
                               'FSW4U','FSW4L','FSW5U','FSW5L','FSW6','FSW7','FSW8'],rotation=20)
plt.xlabel('Coils')
plt.tight_layout()
```
